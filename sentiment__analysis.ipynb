```
import pandas as pd
import numpy as np
```

```
url="https://raw.githubusercontent.com/imsy875/group-by-a/master/sentiment%20analysis/train.csv"
```

```
url2="https://raw.githubusercontent.com/imsy875/group-by-a/master/sentiment%20analysis/test.csv"
```

```
test=pd.read_csv(url2)
```

```
test
```

```
train=pd.read_csv(url)
train
```

```
master=pd.concat([train,test])
```

```
master
```

```
master=master.set_index("id")
```

```
master
```

```
master["tweet"].replace("@\w+", "", regex=True, inplace=True)
```

```
master["tweet"].replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
senti_countvectorizer=vectorizer.fit_transform(master['tweet'])
```

```
encod=pd.DataFrame(senti_countvectorizer.toarray())
```

```
encod
```

```
sentiments=encod.iloc[0:7920,:]
sentiments
```

```
test_encod=encod.iloc[7920:,:]
```

```
test_encod
```

```
train
```

```
y_train=train["label"]
```

```
y_train
```

```
sentiments
```

```
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier(n_estimators=100, criterion = 'entropy')
```

```
random_forest.fit(sentiments,y_train)
```

```
y_pred=random_forest.predict(test_encod)
```

```
y_pred
```

```
test.columns
```

```
test=test.set_index("id")
```

```
test.drop("tweet",axis=1, inplace=True)
```

```
test['label']=y_pred
```

```
test
```

```
test=test.reset_index()
test
```

```
test.to_csv(" try.csv", index= False)
```
